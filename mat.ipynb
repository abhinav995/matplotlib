```python
month_list=df["month_number"]
x=df["toothpaste"]
y=df["bathingsoap"]
width=0.50
plt.bar([a+0.50 for a in month_list],x,label=x,width=0.50,align="edge")
plt.bar([a-0.50 for a in month_list],y,label=y,width=-0.50,align="edge")
plt.xlabel("month number")
plt.ylabel("sales unit")
plt.xticks(month_list)
plt.grid()
plt.show()

    
    
```

```python
import matplotlib.pyplot as plt
import pandas as pd


month_list = df["month_number"]
x = df["toothpaste"]
y = df["bathingsoap"]

width = 0.35
plt.bar(month_list - width/2, x, label='Toothpaste',width=width)
plt.bar(month_list + width/2, y, label='Bathing Soap',width=width)

plt.xlabel("Month Number")
plt.ylabel("Sales Unit")
plt.xticks(month_list)

plt.legend()
plt.show()
```

```python
mnth=df["month_number"]
facecream_data=df["facecream"]
facewash_data=df["facewash"]
toothpaste_data=df["toothpaste"]
bathingsoap_data=df["bathingsoap"]
shampoo_data=df["shampoo"]
moisturizer_data=df["moisturizer"]
plt.plot([],[],color="r",label="facecream",linewidth=5)
plt.plot([],[],color="b",label="facewash",linewidth=5)
plt.plot([],[],color="k",label="toothpaste",linewidth=5)
plt.plot([],[],color="y",label="bathingsoap",linewidth=5)
plt.plot([],[],color="g",label="shampoo",linewidth=5)
plt.plot([],[],color="c",label="moisturizer",linewidth=5)
plt.stackplot(mnth,facecream_data,facewash_data,toothpaste_data,bathingsoap_data,shampoo_data,moisturizer_data,colors=["r","b","k","y","g","c"])
plt.show()
```
